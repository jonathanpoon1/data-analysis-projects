```python
#this file is where I tested most of the code on primarily the first Pokemon (Bulbasaur)

from bs4 import BeautifulSoup
import requests
```

```python
pokemon = 'Bulbasaur'
pokemon2 = 'Mr-Mime'
pokemon3 = 'Rotom'

url = 'https://pokemondb.net/pokedex/' + pokemon

page = requests.get(url)
soup = BeautifulSoup(page.text, 'html')

#info taken (type) will only be based on the first (original) form given

forms_table = soup.find_all('div', class_ = 'sv-tabs-tab-list')[0]
print(forms_table)
```

```python
forms = forms_table.find_all('a')
forms_count = len(forms)
forms_count
```

```python
table = soup.find_all('div', class_ = 'grid-col span-md-12 span-lg-8')[forms_count]
print (table)
```

```python
for br in table.find_all("br"):
    br.replace_with("\n")

games = table.find_all('th')
games
```

```python

games2 = [title.text.splitlines() for title in games]
print(games2)

games3 = [item for sublist in games2 for item in sublist]
games3.insert(0,'Name')
headers = games3
print(games3)

```

```python
import pandas as pd
df = pd.DataFrame(columns = headers)

row_data = table.find_all('tr')
row_data
```

```python
values = [None] * 38
dictionary = dict(zip(games3, values))

test = [title.text.strip() for title in row_data]

print(test)


test2 = []
i = 0
for x in test:
    newlines = len(x.splitlines()) - 1

    for y in range(newlines):
        pair = {x.splitlines()[y]:x.splitlines()[newlines]}
        test2.append(pair)
        i += 1

print(test2)

for x in test2:
    dictionary.update(x)
    
pair = {'Name':pokemon}
dictionary.update(pair)
print(dictionary)
```

```python
table2 = soup.find_all('table', class_ = 'vitals-table')[0]
table2

type1 = table2.find_all('a')[0].text
type2 = table2.find_all('a')[1].text
print(type1)
print(type2)

types_list = ['Normal','Fire','Water','Electric','Grass','Ice','Fighting','Poison','Ground','Flying','Psychic','Bug','Rock','Ghost','Dragon','Dark','Steel','Fairy']

if type2 not in types_list:
    type2 = 'None'

pair1 = {'Type1': type1}
pair2 = {'Type2': type2}
dictionary.update(pair1)
dictionary.update(pair2)
```

```python
test_df = pd.DataFrame([dictionary])

df = pd.concat([df, test_df], ignore_index=True)
names = df.pop("Name")
df.insert(0,"Name", names)
type1s = df.pop("Type1")
df.insert(1,"Type1", type1s)
type2s = df.pop("Type2")
df.insert(2,"Type2", type2s)

df
```

```python
df = pd.read_csv('data.csv')

#replace "Trade/migrate from another game", "Not available in this game", and Na values with 0

df.replace('Trade/migrate from another game', 0, inplace=True)
df.replace('Not available in this game', 0, inplace=True)
df.fillna(0, inplace=True)

#replace all non-zero values in columns except first three with 1s 

games3.remove("Name")

for game in games3:
    df[game] = df[game].where(df[game] == 0, 1)

df.to_csv(r'C:\Users\jonat\Documents\to website\web scraping project (pokemon)\finaldata.csv', index = False)
df
```

```python
#create second dataframe to hold type totals

if "Type" not in games3:
    games3.insert(0,'Type')

print(games3)
new_df = pd.DataFrame(columns = games3)
if "Type" in games3:
    games3.remove('Type')

if "None" not in types_list:
    types_list.append('None')

for type in types_list:
    #go through original df, if type is in type1 or type2, add data
    new_row = []
    new_row.append(type)
    for game in games3:
        total = 0
        for index, row in df.iterrows():
            if row['Type1'] == type or row['Type2'] == type:
                total += row[game]     
        new_row.append(total)
    print(new_row)
    length = len(new_df)
    new_df.loc[length] = new_row

#want a total row at the end where it shows the total number of pokemon available in each game

total_row = []
total_row.append ("Total")
for game in games3:
    total = 0
    for index, row in df.iterrows():
        total += row[game]
    total_row.append(total)

length = len(new_df)
new_df.loc[length] = total_row
print(total_row)

new_df.to_csv(r'C:\Users\jonat\Documents\to website\web scraping project (pokemon)\type_total_by_game.csv', index = False)
new_df
```
